# Phase 1: Foundation & Data Prep
### The Goal: Prepare clean, numerical data that the machine can understand.

**Technical Explanation:**
1. **Mapping:** Machines don't understand "M" (Malignant) or "B" (Benign). We must convert them to 1 and 0.
2. **Dropping:** Columns like ID are unique identifiers; they have no predictive power and must be removed to prevent overfitting.
3. **Scaling:** Essential for distance-based algorithms (like SVM/SGD) to ensure all features are treated equally.

```python
# Step 1: Data Ingestion & Cleaning
import pandas as pd
import numpy as np

# Load Data
# Note: Adjust the path to where your data.csv is actually located
df = pd.read_csv('../../../data/row/data.csv') 

# Mapping: Convert Diagnosis to Binary (1 = Malignant/Cancer, 0 = Benign/Safe)
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})

# Dropping: Remove 'noise' columns that don't help prediction
cols_to_drop = ['id', 'Unnamed: 32']
df = df.drop(columns=[c for c in cols_to_drop if c in df.columns])

# Define Features (X) and Target (y)
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']

print("‚úÖ Data Loaded & Cleaned")
print(f"Dimensions: {X.shape}")
```

```python
# CHECKING FOR IMBALANCE (Crucial Step)
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the percentage of Benign vs Malignant
counts = df['diagnosis'].value_counts()
percentages = df['diagnosis'].value_counts(normalize=True) * 100

print("üìä Class Distribution:")
print(f"Benign (0): {counts[0]} ({percentages[0]:.1f}%)")
print(f"Malignant (1): {counts[1]} ({percentages[1]:.1f}%)")

# Visualization
plt.figure(figsize=(6, 4))
sns.countplot(x='diagnosis', data=df, palette=['green', 'red'])
plt.xticks([0, 1], ['Benign (0)', 'Malignant (1)'])
plt.title('Class Balance Check')
plt.show()

# Logical Conclusion for the Notebook
if percentages[1] < 40:
    print("‚ö†Ô∏è DATA IS IMBALANCED!")
    print("   Action Plan: We must use 'class_weight=balanced' in our models")
    print("   to tell the AI to pay extra attention to the minority class (Cancer).")
else:
    print("‚úÖ Data is relatively balanced. Standard training should work fine.")
```

```python
# Step 2: Splitting & Scaling (Crucial for SVM)
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Split: Hide 30% of data for testing (Simulating new patients)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scaling: Essential for SVM/SGD to treat all features equally
scaler = StandardScaler()

# Fit ONLY on training data (to avoid cheating/leakage)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("‚úÖ Data Split & Scaled")
```

# Phase 2: The Engine (Modeling & Comparison)
### The Goal: Train multiple algorithms to find the best "Brain" for our triage system.

We will not rely on a single algorithm. Instead, we will train three distinct models, ranging from simple to complex, and pit them against each other to ensure we select the most accurate and robust solution for the hospital.

**The Contenders:**
1.  **SGD-SVM (The Efficient Linear):**
    *   *Mechanism:* Uses Stochastic Gradient Descent to find the widest margin (Hinge Loss) between classes.
    *   *Strength:* Fast, scalable, and highly interpretable.
2.  **Logistic Regression (The Medical Standard):**
    *   *Mechanism:* Probabilistic linear classifier commonly used in clinical settings.
    *   *Strength:* Provides native probabilities and is the baseline for medical research.
3.  **XGBoost (The Non-Linear Powerhouse):**
    *   *Mechanism:* Gradient Boosting framework that builds an ensemble of weak prediction trees.
    *   *Strength:* Captures complex, non-linear relationships that linear models might miss.

**Evaluation Strategy:**
We will train all three, tune their hyperparameters, and finally compare them using the **ROC Curve (AUC)** to declare a winner.

```python
# Step 3: Model Training (SGD Classifier with Batch Settings)
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

print("\n" + "=" * 70)
print("SGD CLASSIFIER WITH SVM LOSS (BATCH TRAINING)")
print("=" * 70)

# Calculate dynamic alpha based on training set size
# alpha = 1/(C * n_samples) -> Here C=5 is implied by your formula
dynamic_alpha = 1 / (5 * len(y_train))

# Initialize with your Specific Parameters
model = SGDClassifier(
    loss='hinge', 
    penalty='l2', 
    alpha=dynamic_alpha,
    class_weight='balanced',  # <--- THIS IS THE NEW ADDITION
    learning_rate='constant',
    eta0=1e-3, 
    max_iter=3000, 
    tol=1e-4,
    random_state=42,
    n_jobs=-1
)

print(f"üöÄ Training SGD SVM with alpha={dynamic_alpha:.6f}...")
model.fit(X_train_scaled, y_train)

# Evaluate
y_pred_sgd = model.predict(X_test_scaled)

metrics_sgd = {
    'Accuracy': accuracy_score(y_test, y_pred_sgd),
    'Sensitivity': recall_score(y_test, y_pred_sgd),
    'Precision': precision_score(y_test, y_pred_sgd),
    'F1-Score': f1_score(y_test, y_pred_sgd)
}

print("\nüìä SGD SVM PERFORMANCE:")
for metric, value in metrics_sgd.items():
    print(f"   {metric}: {value:.4f}")
```

```python
# Model 2: Logistic Regression (Baseline)
from sklearn.linear_model import LogisticRegression

print("=" * 70)
print("LOGISTIC REGRESSION (BASELINE)")
print("=" * 70)

# Initialize with Class Weight Balanced
lr_model = LogisticRegression(
    max_iter=10000, 
    class_weight='balanced', # <--- THIS IS THE NEW ADDITION
    random_state=42
)
lr_model.fit(X_train_scaled, y_train)

# Get Probabilities
y_probs_lr = lr_model.predict_proba(X_test_scaled)[:, 1]

print("‚úÖ Logistic Regression Training Completed!")
```

```python
# Model 3: XGBoost with Hyperparameter Tuning
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from scipy.stats import randint, uniform

print("=" * 70)
print("XGBOOST HYPERPARAMETER TUNING")
print("=" * 70)

# Define parameter distributions
param_distributions = {
    'max_depth': randint(3, 12),
    'learning_rate': uniform(0.01, 0.3),
    'n_estimators': randint(100, 500),
    'subsample': uniform(0.6, 0.4),
    'colsample_bytree': uniform(0.6, 0.4),
    'scale_pos_weight': [1, 1.5, 2, 2.5, 3]
}

# Create base model
base_xgb = xgb.XGBClassifier(
    random_state=42,
    eval_metric='logloss',
    use_label_encoder=False
)

# Randomized search
xgb_random_search = RandomizedSearchCV(
    estimator=base_xgb,
    param_distributions=param_distributions,
    n_iter=50,
    scoring='recall',
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42),
    n_jobs=-1,
    random_state=42,
    verbose=1
)

print("Training XGBoost with randomized search...")
# Note: XGBoost handles scaling internally, but for fair comparison we use X_train_scaled
xgb_random_search.fit(X_train_scaled, y_train)

# Get Best Model & Probabilities
best_xgb = xgb_random_search.best_estimator_
y_probs_xgb = best_xgb.predict_proba(X_test_scaled)[:, 1]

print("‚úÖ XGBoost Training Completed!")
print(f"Best Parameters: {xgb_random_search.best_params_}")
```

```python
# Visualizing the Battle: SMOOTHED ROC Curve Analysis with Zoom
from sklearn.metrics import roc_curve, auc
from scipy.interpolate import make_interp_spline
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

def plot_smooth_roc(ax, y_true, y_probs, label, color, linestyle='-'):
    # 1. Calculate actual jagged ROC
    fpr, tpr, _ = roc_curve(y_true, y_probs)
    roc_auc = auc(fpr, tpr)
    
    # 2. Smoothing Logic (Interpolation)
    x_new = np.linspace(0, 1, 300) 
    sorted_indices = np.argsort(fpr)
    fpr_sorted = fpr[sorted_indices]
    tpr_sorted = tpr[sorted_indices]
    
    if len(fpr_sorted) > 3:
        try:
            spl = make_interp_spline(fpr_sorted, tpr_sorted, k=3)
            tpr_smooth = spl(x_new)
            tpr_smooth = np.clip(tpr_smooth, 0, 1)
            ax.plot(x_new, tpr_smooth, color=color, linestyle=linestyle, 
                     linewidth=2, label=f'{label} (AUC = {roc_auc:.4f})')
        except:
            ax.plot(fpr, tpr, color=color, linestyle=linestyle, 
                     linewidth=2, label=f'{label} (AUC = {roc_auc:.4f})')
    else:
        ax.plot(fpr, tpr, color=color, linestyle=linestyle, 
                 linewidth=2, label=f'{label} (AUC = {roc_auc:.4f})')
    return roc_auc

# Create Main Plot
fig, ax = plt.subplots(figsize=(12, 8))

# 1. SGD-SVM (Calibrated) - Green
auc_sgd = plot_smooth_roc(ax, y_test, y_probs, 'SGD-SVM', 'green')

# 2. Logistic Regression - Blue
auc_lr = plot_smooth_roc(ax, y_test, y_probs_lr, 'Logistic Regression', 'blue', '--')

# 3. XGBoost - Red
auc_xgb = plot_smooth_roc(ax, y_test, y_probs_xgb, 'XGBoost', 'red')

# Main Plot Details
ax.plot([0, 1], [0, 1], 'k--', alpha=0.4, linewidth=1) 
ax.set_xlabel('False Positive Rate (False Alarms)', fontsize=12)
ax.set_ylabel('True Positive Rate (Sensitivity)', fontsize=12)
ax.set_title('Model Comparison: Smoothed ROC Curves', fontsize=15, fontweight='bold')
ax.legend(loc='lower right', fontsize=12, frameon=True, shadow=True)
ax.grid(True, alpha=0.2)

# --- ZOOM INSET START ---
# Create an inset plot inside the main plot
# loc=center right means putting the zoom box in the middle right area
axins = inset_axes(ax, width="40%", height="40%", loc='center right', borderpad=3)

# Plot the same data on the inset
plot_smooth_roc(axins, y_test, y_probs, 'SGD-SVM', 'green')
plot_smooth_roc(axins, y_test, y_probs_lr, 'Logistic Regression', 'blue', '--')
plot_smooth_roc(axins, y_test, y_probs_xgb, 'XGBoost', 'red')

# Set the zoom limits (FOCUS ON TOP LEFT CORNER)
# Adjust these values if needed to see better
axins.set_xlim(0.0, 0.15)  # Zoom on X axis (False Positives 0% to 15%)
axins.set_ylim(0.90, 1.01) # Zoom on Y axis (Sensitivity 90% to 100%)

# Add grid to zoom and remove labels to keep it clean
axins.grid(True, alpha=0.2)
axins.set_title("Zoom: Critical Area", fontsize=10)
# --- ZOOM INSET END ---

plt.show()

# Decision Logic
models = {'SGD-SVM': auc_sgd, 'Logistic': auc_lr, 'XGBoost': auc_xgb}
winner = max(models, key=models.get)
print(f"üèÜ AND THE WINNER IS: {winner} with AUC {models[winner]:.4f}")
```

### üìñ Understanding the Metrics (Medical Context)

**1. False Negatives (FN) - The "Danger" Metric üö®**
*   **What it means:** The model said "Safe", but the patient actually had **Cancer**.
*   **Why it matters:** This is the worst-case scenario. We sent a sick patient home.
*   **Goal:** This number must be as close to **0** as possible.

**2. False Positives (FP) - The "Anxiety" Metric üòü**
*   **What it means:** The model said "Cancer", but the patient was actually **Safe**.
*   **Why it matters:** Causes unnecessary stress and expensive biopsies.
*   **Goal:** Keep it low, but it's better to have a False Alarm than to miss a cancer.

**3. Sensitivity (Recall) - The "Safety Net" üï∏Ô∏è**
*   **Formula:** $TP / (TP + FN)$
*   **Logic:** Out of all the sick people, how many did we find?
*   **Target:** > 95%. A high sensitivity means we rarely miss a case.

**4. Precision - The "Trust" Metric üéØ**
*   **Formula:** $TP / (TP + FP)$
*   **Logic:** When the AI says "Cancer", how often is it right?
*   **Target:** High precision means doctors trust the AI's alerts.


```python
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score

# Define a function to calculate metrics for any model
def get_clinical_metrics(model_name, y_true, y_pred):
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    
    return {
        'Model': model_name,
        'True Positives (Caught Cancer)': tp,
        'False Negatives (MISSED Cancer)': fn,  # CRITICAL in Medicine
        'False Positives (False Alarm)': fp,    # Causes anxiety/cost
        'True Negatives (Correctly Safe)': tn,
        'Sensitivity (Recall)': recall_score(y_true, y_pred),
        'Precision': precision_score(y_true, y_pred),
        'F1 Score': f1_score(y_true, y_pred)
    }

# 1. Get Predictions (we need binary 0/1, not probabilities here)
y_pred_sgd = model.predict(X_test_scaled)
y_pred_lr = lr_model.predict(X_test_scaled)
y_pred_xgb = best_xgb.predict(X_test_scaled)

# 2. Generate Report
results = []
results.append(get_clinical_metrics('SGD-SVM', y_test, y_pred_sgd))
results.append(get_clinical_metrics('Logistic Reg', y_test, y_pred_lr))
results.append(get_clinical_metrics('XGBoost', y_test, y_pred_xgb))

# 3. Create DataFrame
clinical_df = pd.DataFrame(results).set_index('Model')

# 4. Display
print("\nüè• CLINICAL PERFORMANCE REPORT")
print("=" * 80)
display(clinical_df)
```

# Phase 4: Industrialization
### The Goal: Make the code clean, reusable, and ready for deployment.

**Strategy:**
* Wrap logic into **Python Classes** (OOP).
* Create a **Pipeline** that can handle new data automatically.

```python
# ---------------------------------------------------------
# STEP 7: INDUSTRIALIZED PIPELINE & DASHBOARD SIMULATION
# ---------------------------------------------------------
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.calibration import CalibratedClassifierCV
from sklearn.linear_model import SGDClassifier

# --- 1. CLASS DEFINITIONS ---

class DataHandler:
    def __init__(self, filepath):
        self.filepath = filepath
    def load_and_clean(self):
        df = pd.read_csv(self.filepath)
        df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
        df = df.drop(['id', 'Unnamed: 32'], axis=1, errors='ignore')
        return df.drop('diagnosis', axis=1), df['diagnosis']

class FeatureEngineer:
    def __init__(self):
        self.scaler = StandardScaler()
    def fit_transform(self, X):
        return self.scaler.fit_transform(X)
    def transform(self, X):
        return self.scaler.transform(X)

class CancerModel:
    def __init__(self):
        self.model = None
    def train(self, X_train, y_train):
        base_model = SGDClassifier(loss='hinge', penalty='l2', alpha=0.01, 
                                   random_state=42, class_weight='balanced')
        self.model = CalibratedClassifierCV(base_model, method='sigmoid', cv=5)
        self.model.fit(X_train, y_train)
        print("‚úÖ Model Trained & Calibrated.")
    def predict_proba(self, X_data):
        return self.model.predict_proba(X_data)

class RiskStratifier:
    def __init__(self):
        # Dictionary to map raw columns to Human Readable + Units
        self.feature_map = {
            'radius_mean': 'Radius Mean (mm)', 'texture_mean': 'Texture Mean (gray-scale)', 
            'perimeter_mean': 'Perimeter Mean (mm)', 'area_mean': 'Area Mean (mm¬≤)', 
            'smoothness_mean': 'Smoothness Mean (local var)', 'compactness_mean': 'Compactness Mean', 
            'concavity_mean': 'Concavity Mean', 'concave points_mean': 'Concave Pts Mean', 
            'symmetry_mean': 'Symmetry Mean', 'fractal_dimension_mean': 'Fractal Dim Mean',
            'radius_se': 'Radius SE', 'texture_se': 'Texture SE', 'perimeter_se': 'Perimeter SE', 
            'area_se': 'Area SE', 'smoothness_se': 'Smoothness SE', 'compactness_se': 'Compactness SE', 
            'concavity_se': 'Concavity SE', 'concave points_se': 'Concave Pts SE', 
            'symmetry_se': 'Symmetry SE', 'fractal_dimension_se': 'Fractal Dim SE',
            'radius_worst': 'Worst Radius (mm)', 'texture_worst': 'Worst Texture (gray-scale)', 
            'perimeter_worst': 'Worst Perimeter (mm)', 'area_worst': 'Worst Area (mm¬≤)', 
            'smoothness_worst': 'Worst Smoothness', 'compactness_worst': 'Worst Compactness', 
            'concavity_worst': 'Worst Concavity', 'concave points_worst': 'Worst Concave Pts', 
            'symmetry_worst': 'Worst Symmetry', 'fractal_dimension_worst': 'Worst Fractal Dim'
        }

    def get_triage_data(self, prob, features_series_scaled, features_series_raw, feature_names):
        # 1. Determine Status & Action
        if prob >= 0.80:
            status = "üî¥ Critical"
            action = "Book Biopsy Now"
        elif prob >= 0.40:
            status = "üü† Urgent"
            action = "Review Today"
        else:
            status = "üü¢ Low"
            action = "Auto-scheduled 6mo"
            
        # 2. Determine Top 3 Key Factors
        # We identify factors based on the magnitude of the SCALED value (deviation from norm)
        # But we DISPLAY the RAW value for medical context.
        
        # Sort indices by absolute scaled value (biggest outliers first)
        top_indices = np.argsort(np.abs(features_series_scaled))[::-1][:3]
        
        factor_strings = []
        for idx in top_indices:
            raw_name = feature_names[idx]
            raw_val = features_series_raw.iloc[idx] # Access raw value by index
            
            # Get readable name with unit, default to raw name if not found
            readable_name = self.feature_map.get(raw_name, raw_name)
            
            # Format string: "Name (Value)"
            factor_strings.append(f"{readable_name}: {raw_val:.2f}")
            
        # Join with commas
        factors_display = ", ".join(factor_strings)
            
        return status, factors_display, action

# --- 2. EXECUTION & DASHBOARD GENERATION ---

if __name__ == "__main__":
    print("\nüè• HOSPITAL TRIAGE SYSTEM - BATCH PROCESSING")
    print("=" * 80)
    
    # 1. Initialize
    dh = DataHandler('../../../data/row/data.csv') 
    X, y = dh.load_and_clean()
    X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.3, random_state=42)
    
    fe = FeatureEngineer()
    X_tr_s = fe.fit_transform(X_tr)
    X_te_s = fe.transform(X_te)
    
    cm = CancerModel()
    cm.train(X_tr_s, y_tr)
    rs = RiskStratifier()
    
    # 2. Simulate Batch
    batch_size = 5
    random_indices = np.random.choice(len(X_te), batch_size, replace=False)
    
    dashboard_data = []
    
    for idx in random_indices:
        # Get Model Input (Scaled)
        patient_scaled_arr = X_te_s[idx]
        patient_scaled_reshaped = patient_scaled_arr.reshape(1, -1)
        
        # Get Raw Input (For Display) - Access via iloc using the random index
        # Note: X_te is a DataFrame, so we use iloc[idx] relative to the test set? 
        # CAUTION: random_indices are 0..len(X_te), so X_te.iloc[idx] is correct.
        patient_raw_series = X_te.iloc[idx]
        
        # Predict
        prob = cm.predict_proba(patient_scaled_reshaped)[0][1]
        
        # Get Triage Logic (Pass both scaled and raw data)
        status, factors, action = rs.get_triage_data(prob, patient_scaled_arr, patient_raw_series, X.columns)
        
        dashboard_data.append({
            "Patient ID": f"P-{9000 + idx}", 
            "Risk Score": f"{prob:.2f}", 
            "Status": status, 
            "Key Factors (Top 3 Outliers)": factors, 
            "Action Required": action
        })
    
    # 3. Create DataFrame
    dashboard_df = pd.DataFrame(dashboard_data)
    
    # 4. Styling
    print(f"Incoming Scans (Batch: #{np.random.randint(1000,9999)}) {' '*40} Processed: Just now")
    
    def color_rows(val):
        s = str(val)
        if 'Critical' in s: return 'color: red; font-weight: bold;'
        if 'Urgent' in s: return 'color: #e67e22; font-weight: bold;'
        if 'Low' in s: return 'color: #27ae60;'
        if 'Book' in s: return 'background-color: #c0392b; color: white; border-radius: 4px; padding: 2px 6px;'
        if 'Review' in s: return 'border: 1px solid #e67e22; color: #e67e22; border-radius: 4px; padding: 2px 6px;'
        return ''

    styler = dashboard_df.style\
        .applymap(lambda x: color_rows(x) if any(k in str(x) for k in ['Critical', 'Urgent', 'Low']) else '', subset=['Status', 'Risk Score'])\
        .applymap(lambda x: color_rows(x) if any(k in str(x) for k in ['Book', 'Review']) else '', subset=['Action Required'])\
        .set_properties(**{'text-align': 'left', 'padding': '8px', 'font-family': 'sans-serif'})

    # Safe Hide Index
    try: styler = styler.hide(axis="index")
    except: 
        try: styler = styler.hide_index()
        except: pass
            
    display(styler)
```
