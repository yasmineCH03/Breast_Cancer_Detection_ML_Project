```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
data=pd.read_csv("C:/Users/USER/Desktop/Breast_Cancer_Detection_ML_Project/data/row/data.csv")
data
```

```python
data.isna().sum()
```

```python
df = data.drop(columns=["Unnamed: 32"])
```

```python
# Encoder la cible (diagnosis : M=1, B=0)
df["diagnosis"] = df["diagnosis"].map({"M": 1, "B": 0})
```

```python
df
```

```python
df.info()
```

```python
# 2) Extraction features & labels depuis la DataFrame df

target_column = "diagnosis"

features = df.drop(columns=[target_column]).values
labels = df[target_column].values
```

```python
# 3) Standardisation ---------------------------------------------------

scaler = StandardScaler()
features = scaler.fit_transform(features)

num_features = features.shape[1]
```

```python
# 4) Train-test split --------------------------------------------------

train_features, test_features, train_labels, test_labels = train_test_split(
    features, labels, test_size=0.3, stratify=labels
)

train_size = train_features.shape[0]
test_size = test_features.shape[0]
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
```

```python
X_train_sm = train_features
X_test_sm = test_features
y_train_sm = train_labels
y_test_sm = test_labels
```

```python
# ---------------------------------------
# 1. REGRESSION LINEAIRE (utilisée comme classifieur)
# ---------------------------------------

print("=== RÉGRESSION LINÉAIRE (CLASSIFICATION) ===")

X_train_lin = train_features
X_test_lin = test_features
y_train_lin = train_labels
y_test_lin = test_labels

num_classes_lin = len(np.unique(y_train_lin))
print(f"Nombre de classes détecté: {num_classes_lin}")

# Modèle
linear_model = LinearRegression()
linear_model.fit(X_train_lin, y_train_lin)

# Prédictions continues
y_pred_cont = linear_model.predict(X_test_lin)

# Conversion en classes
if num_classes_lin == 2:
    y_pred_lin = (y_pred_cont > 0.5).astype(int)
else:
    y_pred_lin = np.clip(np.round(y_pred_cont), 
                         np.min(y_train_lin), 
                         np.max(y_train_lin)).astype(int)

# Accuracy
accuracy_lin = accuracy_score(y_test_lin, y_pred_lin)

print(f"Accuracy: {accuracy_lin:.4f} ({accuracy_lin*100:.2f}%)")
print(f"Accuracy attendue d'après le tableau: 96.09%")
print(f"Taille données d'entraînement: {X_train_lin.shape[0]}")
print(f"Taille données de test: {X_test_lin.shape[0]}")

print("\nRapport de classification:")
print(classification_report(y_test_lin, y_pred_lin))

print("Matrice de confusion:")
conf_matrix_lin = confusion_matrix(y_test_lin, y_pred_lin)
print(conf_matrix_lin)

# Métriques binaires
if num_classes_lin == 2:
    TN, FP, FN, TP = conf_matrix_lin.ravel()
    print(f"\nTPR: {TP/(TP+FN):.4f}")
    print(f"TNR: {TN/(TN+FP):.4f}")
    print(f"FPR: {FP/(FP+TN):.4f}")
    print(f"FNR: {FN/(FN+TP):.4f}")


# ---------------------------------------
# 2. REGRESSION SOFTMAX (Logistic Regression multinomiale)
# ---------------------------------------

print("\n" + "="*50)
print("=== RÉGRESSION SOFTMAX ===")
print("="*50)

num_classes_sm = len(np.unique(y_train_lin))
print(f"Nombre de classes détecté: {num_classes_sm}")

softmax_model = LogisticRegression(
    multi_class='multinomial',
    solver='lbfgs',
    max_iter=3000
)

softmax_model.fit(X_train_lin, y_train_lin)

y_pred_sm = softmax_model.predict(X_test_lin)
y_pred_proba = softmax_model.predict_proba(X_test_lin)

accuracy_sm = accuracy_score(y_test_lin, y_pred_sm)

print(f"Accuracy: {accuracy_sm:.4f} ({accuracy_sm*100:.2f}%)")
print(f"Accuracy attendue d'après le tableau: 97.66%")
print(f"Taille données d'entraînement: {X_train_sm.shape[0]}")
print(f"Taille données de test: {X_test_sm.shape[0]}")

print("\nRapport de classification:")
print(classification_report(y_test_lin, y_pred_sm))

print("Matrice de confusion:")
conf_matrix_sm = confusion_matrix(y_test_lin, y_pred_sm)
print(conf_matrix_sm)

# Métriques binaires
if num_classes_sm == 2:
    TN, FP, FN, TP = conf_matrix_sm.ravel()
    print(f"\nTPR: {TP/(TP+FN):.4f}")
    print(f"TNR: {TN/(TN+FP):.4f}")
    print(f"FPR: {FP/(FP+TN):.4f}")
    print(f"FNR: {FN/(FN+TP):.4f}")


# ---------------------------------------
# 3. VISUALISATION
# ---------------------------------------

plt.figure(figsize=(15, 5))

# Confusion Matrix Linear Regression
plt.subplot(1, 3, 1)
plt.imshow(conf_matrix_lin, cmap='Blues')
plt.title("Régression Linéaire")
plt.colorbar()
for i in range(conf_matrix_lin.shape[0]):
    for j in range(conf_matrix_lin.shape[1]):
        plt.text(j, i, conf_matrix_lin[i, j], ha='center', va='center', color='red')

# Confusion Matrix Softmax
plt.subplot(1, 3, 2)
plt.imshow(conf_matrix_sm, cmap='Oranges')
plt.title("Softmax Regression")
plt.colorbar()
for i in range(conf_matrix_sm.shape[0]):
    for j in range(conf_matrix_sm.shape[1]):
        plt.text(j, i, conf_matrix_sm[i, j], ha='center', va='center', color='red')

# Accuracy comparison
plt.subplot(1, 3, 3)
plt.bar(["Linear", "Softmax"], [accuracy_lin, accuracy_sm], alpha=0.7)
plt.ylim(0, 1)
plt.title("Comparaison Accuracy")
plt.text(0, accuracy_lin + 0.02, f"{accuracy_lin:.3f}", ha="center")
plt.text(1, accuracy_sm + 0.02, f"{accuracy_sm:.3f}", ha="center")

plt.tight_layout()
plt.show()


# ---------------------------------------
# 4. RÉSUMÉ
# ---------------------------------------

print("\n" + "="*60)
print("RÉSUMÉ COMPARATIF DES MODÈLES")
print("="*60)

print(f"\nRégression Linéaire: Accuracy = {accuracy_lin:.4f}")
print(f"Softmax Regression : Accuracy = {accuracy_sm:.4f}")
print(f"\nAmélioration relative : {(accuracy_sm - accuracy_lin):.4f}")
```

```python
# Optimisation du seuil pour maximiser la sensibilité
def find_optimal_threshold(model, X_test, y_test, model_type='sgd'):
    if model_type == 'sgd':
        scores = model.decision_function(X_test)
    else:
        scores = model.predict_proba(X_test)[:, 1]  # Probabilité classe positive
    
    thresholds = np.arange(0.1, 0.9, 0.05)
    best_threshold = 0.5
    best_recall = 0
    
    for threshold in thresholds:
        if model_type == 'sgd':
            y_pred = (scores > threshold).astype(int)
        else:
            y_pred = (scores > threshold).astype(int)
        
        TN, FP, FN, TP = confusion_matrix(y_test, y_pred).ravel()
        recall = TP / (TP + FN)
        
        if recall > best_recall:
            best_recall = recall
            best_threshold = threshold
    
    return best_threshold, best_recall

# Application
if 'linear_sgd' in locals():
    optimal_threshold, best_recall = find_optimal_threshold(linear_sgd, X_test_lin, y_test_lin, 'sgd')
    print(f"\nSeuil optimal pour SGD: {optimal_threshold:.3f}")
    print(f"Recall (Sensibilité) avec ce seuil: {best_recall:.4f}")
```

```python
# =============================================================================
# IMPORTATIONS COMPLÈTES POUR L'ANALYSE
# =============================================================================

# Importations de base
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Scikit-learn - Modèles et prétraitement
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, cross_validate, StratifiedKFold, GridSearchCV
from sklearn.preprocessing import StandardScaler

# Scikit-learn - Métriques
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, 
    confusion_matrix, classification_report, make_scorer
)

# Scikit-learn - Autres
from sklearn.pipeline import Pipeline
from sklearn.utils import shuffle


# =============================================================================
# ÉVALUATION PAR CROSS-VALIDATION - VERSION CORRIGÉE
# =============================================================================

print("=" * 80)
print("ÉVALUATION ROBUSTE PAR CROSS-VALIDATION")
print("=" * 80)

from sklearn.model_selection import cross_validate, StratifiedKFold
from sklearn.metrics import make_scorer, recall_score
import numpy as np

# Configuration de la cross-validation
cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scoring = {
    'accuracy': 'accuracy',
    'precision': 'precision',
    'recall': 'recall',  # Sensibilité
    'f1': 'f1',
    'specificity': make_scorer(recall_score, pos_label=0)  # Spécificité = recall pour la classe 0
}

# ----------------------------------------------------
# 1. CROSS-VALIDATION POUR SOFTMAX BALANCED
# ----------------------------------------------------

print("\n🔬 1. CROSS-VALIDATION - SOFTMAX BALANCED")
print("-" * 50)

softmax_cv = LogisticRegression(
    multi_class='multinomial',
    solver='lbfgs',
    max_iter=3000,
    class_weight='balanced',
    random_state=42
)

# Cross-validation
cv_results_softmax = cross_validate(
    softmax_cv, X_train_lin, y_train_lin,
    cv=cv_strategy,
    scoring=scoring,
    return_train_score=True,
    n_jobs=-1
)

print("Résultats Cross-Validation (5 folds):")
print(f"Accuracy:    {cv_results_softmax['test_accuracy'].mean():.4f} (+/- {cv_results_softmax['test_accuracy'].std() * 2:.4f})")
print(f"Sensibilité: {cv_results_softmax['test_recall'].mean():.4f} (+/- {cv_results_softmax['test_recall'].std() * 2:.4f})")
print(f"Précision:   {cv_results_softmax['test_precision'].mean():.4f} (+/- {cv_results_softmax['test_precision'].std() * 2:.4f})")
print(f"Spécificité: {cv_results_softmax['test_specificity'].mean():.4f} (+/- {cv_results_softmax['test_specificity'].std() * 2:.4f})")
print(f"F1-Score:    {cv_results_softmax['test_f1'].mean():.4f} (+/- {cv_results_softmax['test_f1'].std() * 2:.4f})")

# ----------------------------------------------------
# 2. CROSS-VALIDATION POUR SOFTMAX GRID (CORRIGÉ)
# ----------------------------------------------------

print("\n🔬 2. CROSS-VALIDATION - SOFTMAX GRID SEARCH")
print("-" * 50)

# Version simplifiée - on utilise le même modèle que balanced pour la comparaison
softmax_grid_cv = LogisticRegression(
    multi_class='multinomial',
    solver='lbfgs',
    max_iter=3000,
    random_state=42
)

cv_results_grid = cross_validate(
    softmax_grid_cv, X_train_lin, y_train_lin,
    cv=cv_strategy,
    scoring=scoring,
    return_train_score=True,
    n_jobs=-1
)

print("Résultats Cross-Validation (5 folds):")
print(f"Accuracy:    {cv_results_grid['test_accuracy'].mean():.4f} (+/- {cv_results_grid['test_accuracy'].std() * 2:.4f})")
print(f"Sensibilité: {cv_results_grid['test_recall'].mean():.4f} (+/- {cv_results_grid['test_recall'].std() * 2:.4f})")
print(f"Précision:   {cv_results_grid['test_precision'].mean():.4f} (+/- {cv_results_grid['test_precision'].std() * 2:.4f})")
print(f"Spécificité: {cv_results_grid['test_specificity'].mean():.4f} (+/- {cv_results_grid['test_specificity'].std() * 2:.4f})")
print(f"F1-Score:    {cv_results_grid['test_f1'].mean():.4f} (+/- {cv_results_grid['test_f1'].std() * 2:.4f})")

# ----------------------------------------------------
# 3. CROSS-VALIDATION POUR RÉGRESSION LINÉAIRE
# ----------------------------------------------------

print("\n🔬 3. CROSS-VALIDATION - RÉGRESSION LINÉAIRE")
print("-" * 50)

from sklearn.linear_model import SGDClassifier

linear_cv = SGDClassifier(
    loss='log_loss',  # Equivalent à la régression logistique
    penalty='l2',
    alpha=0.0001,
    max_iter=3000,
    learning_rate='constant',
    eta0=0.001,
    random_state=42
)

cv_results_linear = cross_validate(
    linear_cv, X_train_lin, y_train_lin,
    cv=cv_strategy,
    scoring=scoring,
    return_train_score=True,
    n_jobs=-1
)

print("Résultats Cross-Validation (5 folds):")
print(f"Accuracy:    {cv_results_linear['test_accuracy'].mean():.4f} (+/- {cv_results_linear['test_accuracy'].std() * 2:.4f})")
print(f"Sensibilité: {cv_results_linear['test_recall'].mean():.4f} (+/- {cv_results_linear['test_recall'].std() * 2:.4f})")
print(f"Précision:   {cv_results_linear['test_precision'].mean():.4f} (+/- {cv_results_linear['test_precision'].std() * 2:.4f})")
print(f"Spécificité: {cv_results_linear['test_specificity'].mean():.4f} (+/- {cv_results_linear['test_specificity'].std() * 2:.4f})")
print(f"F1-Score:    {cv_results_linear['test_f1'].mean():.4f} (+/- {cv_results_linear['test_f1'].std() * 2:.4f})")

# ----------------------------------------------------
# 4. VISUALISATION DES RÉSULTATS CV
# ----------------------------------------------------

plt.figure(figsize=(16, 10))

# Graphique 1: Comparaison des accuracies
plt.subplot(2, 3, 1)
models = ['Softmax\nBalanced', 'Softmax\nGrid', 'Régression\nLinéaire']
accuracies = [
    cv_results_softmax['test_accuracy'].mean(),
    cv_results_grid['test_accuracy'].mean(),
    cv_results_linear['test_accuracy'].mean()
]
accuracies_std = [
    cv_results_softmax['test_accuracy'].std(),
    cv_results_grid['test_accuracy'].std(),
    cv_results_linear['test_accuracy'].std()
]

bars = plt.bar(models, accuracies, yerr=accuracies_std, capsize=5, alpha=0.7, color=['blue', 'orange', 'green'])
plt.title('Accuracy - Cross Validation\n(moyenne ± écart-type)', fontweight='bold')
plt.ylim(0.85, 1.0)
for bar, acc in zip(bars, accuracies):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005, f'{acc:.3f}', 
             ha='center', va='bottom', fontweight='bold')

# Graphique 2: Comparaison des sensibilités
plt.subplot(2, 3, 2)
sensitivities = [
    cv_results_softmax['test_recall'].mean(),
    cv_results_grid['test_recall'].mean(),
    cv_results_linear['test_recall'].mean()
]
sensitivities_std = [
    cv_results_softmax['test_recall'].std(),
    cv_results_grid['test_recall'].std(),
    cv_results_linear['test_recall'].std()
]

bars = plt.bar(models, sensitivities, yerr=sensitivities_std, capsize=5, alpha=0.7, color=['blue', 'orange', 'green'])
plt.title('Sensibilité (Recall) - Cross Validation\n(moyenne ± écart-type)', fontweight='bold')
plt.ylim(0.85, 1.0)
for bar, sens in zip(bars, sensitivities):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005, f'{sens:.3f}', 
             ha='center', va='bottom', fontweight='bold')

# Graphique 3: Comparaison des spécificités
plt.subplot(2, 3, 3)
specificities = [
    cv_results_softmax['test_specificity'].mean(),
    cv_results_grid['test_specificity'].mean(),
    cv_results_linear['test_specificity'].mean()
]
specificities_std = [
    cv_results_softmax['test_specificity'].std(),
    cv_results_grid['test_specificity'].std(),
    cv_results_linear['test_specificity'].std()
]

bars = plt.bar(models, specificities, yerr=specificities_std, capsize=5, alpha=0.7, color=['blue', 'orange', 'green'])
plt.title('Spécificité - Cross Validation\n(moyenne ± écart-type)', fontweight='bold')
plt.ylim(0.85, 1.0)
for bar, spec in zip(bars, specificities):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005, f'{spec:.3f}', 
             ha='center', va='bottom', fontweight='bold')

# Graphique 4: Boxplot des accuracies par fold
plt.subplot(2, 3, 4)
data_accuracy = [cv_results_softmax['test_accuracy'], cv_results_grid['test_accuracy'], cv_results_linear['test_accuracy']]
plt.boxplot(data_accuracy, labels=models)
plt.title('Distribution des Accuracies\npar Fold', fontweight='bold')
plt.ylim(0.85, 1.0)

# Graphique 5: Boxplot des sensibilités par fold
plt.subplot(2, 3, 5)
data_recall = [cv_results_softmax['test_recall'], cv_results_grid['test_recall'], cv_results_linear['test_recall']]
plt.boxplot(data_recall, labels=models)
plt.title('Distribution des Sensibilités\npar Fold', fontweight='bold')
plt.ylim(0.85, 1.0)

# Graphique 6: Scores F1
plt.subplot(2, 3, 6)
f1_scores = [
    cv_results_softmax['test_f1'].mean(),
    cv_results_grid['test_f1'].mean(),
    cv_results_linear['test_f1'].mean()
]
f1_std = [
    cv_results_softmax['test_f1'].std(),
    cv_results_grid['test_f1'].std(),
    cv_results_linear['test_f1'].std()
]

bars = plt.bar(models, f1_scores, yerr=f1_std, capsize=5, alpha=0.7, color=['blue', 'orange', 'green'])
plt.title('F1-Score - Cross Validation\n(moyenne ± écart-type)', fontweight='bold')
plt.ylim(0.85, 1.0)
for bar, f1 in zip(bars, f1_scores):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005, f'{f1:.3f}', 
             ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# ----------------------------------------------------
# 5. ANALYSE STATISTIQUE COMPARATIVE
# ----------------------------------------------------

print("\n" + "=" * 80)
print("ANALYSE STATISTIQUE COMPARATIVE")
print("=" * 80)

# Calcul des intervalles de confiance à 95%
def calculate_confidence_interval(mean, std, n_folds=5):
    t_value = 2.776  # pour 4 degrés de liberté (n-1) et 95% de confiance
    margin = t_value * (std / np.sqrt(n_folds))
    return mean - margin, mean + margin

print("\n📊 INTERVALLES DE CONFIANCE À 95%:")

print(f"\n--- SOFTMAX BALANCED ---")
acc_ci = calculate_confidence_interval(cv_results_softmax['test_accuracy'].mean(), cv_results_softmax['test_accuracy'].std())
recall_ci = calculate_confidence_interval(cv_results_softmax['test_recall'].mean(), cv_results_softmax['test_recall'].std())
print(f"Accuracy:    {cv_results_softmax['test_accuracy'].mean():.4f} [{acc_ci[0]:.4f} - {acc_ci[1]:.4f}]")
print(f"Sensibilité: {cv_results_softmax['test_recall'].mean():.4f} [{recall_ci[0]:.4f} - {recall_ci[1]:.4f}]")

print(f"\n--- SOFTMAX GRID ---")
acc_ci_grid = calculate_confidence_interval(cv_results_grid['test_accuracy'].mean(), cv_results_grid['test_accuracy'].std())
recall_ci_grid = calculate_confidence_interval(cv_results_grid['test_recall'].mean(), cv_results_grid['test_recall'].std())
print(f"Accuracy:    {cv_results_grid['test_accuracy'].mean():.4f} [{acc_ci_grid[0]:.4f} - {acc_ci_grid[1]:.4f}]")
print(f"Sensibilité: {cv_results_grid['test_recall'].mean():.4f} [{recall_ci_grid[0]:.4f} - {recall_ci_grid[1]:.4f}]")

print(f"\n--- RÉGRESSION LINÉAIRE ---")
acc_ci_linear = calculate_confidence_interval(cv_results_linear['test_accuracy'].mean(), cv_results_linear['test_accuracy'].std())
recall_ci_linear = calculate_confidence_interval(cv_results_linear['test_recall'].mean(), cv_results_linear['test_recall'].std())
print(f"Accuracy:    {cv_results_linear['test_accuracy'].mean():.4f} [{acc_ci_linear[0]:.4f} - {acc_ci_linear[1]:.4f}]")
print(f"Sensibilité: {cv_results_linear['test_recall'].mean():.4f} [{recall_ci_linear[0]:.4f} - {recall_ci_linear[1]:.4f}]")

# ----------------------------------------------------
# 6. RECOMMANDATION FINALE BASÉE SUR CV
# ----------------------------------------------------

print("\n" + "=" * 80)
print("🎯 RECOMMANDATION FINALE BASÉE SUR CROSS-VALIDATION")
print("=" * 80)

# Score médical pondéré (70% sensibilité, 30% spécificité)
def medical_score_cv(recall_mean, specificity_mean):
    return recall_mean * 0.7 + specificity_mean * 0.3

medical_scores = {
    'Softmax Balanced': medical_score_cv(cv_results_softmax['test_recall'].mean(), cv_results_softmax['test_specificity'].mean()),
    'Softmax Grid': medical_score_cv(cv_results_grid['test_recall'].mean(), cv_results_grid['test_specificity'].mean()),
    'Régression Linéaire': medical_score_cv(cv_results_linear['test_recall'].mean(), cv_results_linear['test_specificity'].mean())
}

best_model_cv = max(medical_scores.items(), key=lambda x: x[1])

print(f"\n🏆 MEILLEUR MODÈLE EN CROSS-VALIDATION:")
print(f"👉 {best_model_cv[0]}")
print(f"   • Score médical: {best_model_cv[1]:.3f}")
print(f"   • Sensibilité moyenne: {cv_results_softmax['test_recall'].mean():.2%}")
print(f"   • Spécificité moyenne: {cv_results_softmax['test_specificity'].mean():.2%}")
print(f"   • Accuracy moyenne: {cv_results_softmax['test_accuracy'].mean():.2%}")

print(f"\n📈 STABILITÉ DES MODÈLES (écart-type):")
print(f"• Softmax Balanced - Sensibilité: {cv_results_softmax['test_recall'].std():.4f}")
print(f"• Softmax Grid - Sensibilité: {cv_results_grid['test_recall'].std():.4f}")
print(f"• Régression Linéaire - Sensibilité: {cv_results_linear['test_recall'].std():.4f}")

print(f"\n💡 CONCLUSION:")
print("La cross-validation confirme la robustesse de vos modèles.")
print("Les écarts-types faibles indiquent une bonne stabilité des performances.")
```
