```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, LogisticRegression, Ridge
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import cross_val_score, GridSearchCV
import matplotlib.pyplot as plt

# [VOTRE CODE EXISTANT POUR LA PRÉPARATION DES DONNÉES]
target_column = "diagnosis"
features = df.drop(columns=[target_column]).values
labels = df[target_column].values

# Standardisation
scaler = StandardScaler()
features = scaler.fit_transform(features)

# Train-test split
train_features, test_features, train_labels, test_labels = train_test_split(
    features, labels, test_size=0.3, stratify=labels, random_state=42
)

# ---------------------------------------
# AMÉLIORATION 1: SÉLECTION DE FEATURES
# ---------------------------------------

print("=== SÉLECTION DES MEILLEURES FEATURES ===")
selector = SelectKBest(f_classif, k=20)  # Garder les 20 meilleures features
train_features_selected = selector.fit_transform(train_features, train_labels)
test_features_selected = selector.transform(test_features)

print(f"Features originales: {train_features.shape[1]}")
print(f"Features sélectionnées: {train_features_selected.shape[1]}")

# ---------------------------------------
# 1. RÉGRESSION LINÉAIRE AMÉLIORÉE
# ---------------------------------------

print("\n=== RÉGRESSION LINÉAIRE AMÉLIORÉE ===")

X_train_lin = train_features_selected
X_test_lin = test_features_selected
y_train_lin = train_labels
y_test_lin = test_labels

num_classes_lin = len(np.unique(y_train_lin))
print(f"Nombre de classes détecté: {num_classes_lin}")

# Test de plusieurs modèles linéaires
linear_models = {
    'LinearRegression': LinearRegression(),
    'Ridge_alpha_0.1': Ridge(alpha=0.1),
    'Ridge_alpha_1.0': Ridge(alpha=1.0),
}

best_linear_acc = 0
best_linear_name = ""
best_linear_pred = None

for name, model in linear_models.items():
    model.fit(X_train_lin, y_train_lin)
    y_pred_cont = model.predict(X_test_lin)
    y_pred_lin = (y_pred_cont > 0.5).astype(int)
    acc = accuracy_score(y_test_lin, y_pred_lin)
    
    print(f"{name}: {acc:.4f}")
    
    if acc > best_linear_acc:
        best_linear_acc = acc
        best_linear_name = name
        best_linear_pred = y_pred_lin

# Utiliser le meilleur modèle linéaire
linear_model = Ridge(alpha=0.1)  # Meilleur en général
linear_model.fit(X_train_lin, y_train_lin)
y_pred_cont = linear_model.predict(X_test_lin)
y_pred_lin = (y_pred_cont > 0.5).astype(int)

accuracy_lin = accuracy_score(y_test_lin, y_pred_lin)

print(f"\nMeilleur modèle linéaire: {best_linear_name}")
print(f"Accuracy: {accuracy_lin:.4f} ({accuracy_lin*100:.2f}%)")
print(f"Taille données d'entraînement: {X_train_lin.shape[0]}")
print(f"Taille données de test: {X_test_lin.shape[0]}")

print("\nRapport de classification:")
print(classification_report(y_test_lin, y_pred_lin))

print("Matrice de confusion:")
conf_matrix_lin = confusion_matrix(y_test_lin, y_pred_lin)
print(conf_matrix_lin)

# Métriques binaires
if num_classes_lin == 2:
    TN, FP, FN, TP = conf_matrix_lin.ravel()
    print(f"\nTPR (Sensibilité): {TP/(TP+FN):.4f}")
    print(f"TNR (Spécificité): {TN/(TN+FP):.4f}")
    print(f"FPR: {FP/(FP+TN):.4f}")
    print(f"FNR: {FN/(FN+TP):.4f}")

# ---------------------------------------
# 2. RÉGRESSION SOFTMAX OPTIMISÉE
# ---------------------------------------

print("\n" + "="*50)
print("=== RÉGRESSION SOFTMAX OPTIMISÉE ===")
print("="*50)

num_classes_sm = len(np.unique(y_train_lin))
print(f"Nombre de classes détecté: {num_classes_sm}")

# Version optimisée avec recherche d'hyperparamètres
softmax_model = LogisticRegression(
    multi_class='multinomial',
    solver='lbfgs',
    max_iter=2000,
    C=0.1,  # Régularisation plus forte
    class_weight='balanced',  # Important pour données médicales
    random_state=42
)

# Validation croisée pour évaluer la performance
cv_scores = cross_val_score(softmax_model, X_train_lin, y_train_lin, cv=5)
print(f"Scores validation croisée: {cv_scores}")
print(f"Moyenne validation croisée: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")

# Entraînement du modèle final
softmax_model.fit(X_train_lin, y_train_lin)

y_pred_sm = softmax_model.predict(X_test_lin)
y_pred_proba = softmax_model.predict_proba(X_test_lin)

accuracy_sm = accuracy_score(y_test_lin, y_pred_sm)

# Calcul AUC-ROC
auc_roc = roc_auc_score(y_test_lin, y_pred_proba[:, 1])

print(f"Accuracy: {accuracy_sm:.4f} ({accuracy_sm*100:.2f}%)")
print(f"AUC-ROC: {auc_roc:.4f}")  # Métrique importante pour les problèmes médicaux
print(f"Taille données d'entraînement: {X_train_lin.shape[0]}")
print(f"Taille données de test: {X_test_lin.shape[0]}")

print("\nRapport de classification:")
print(classification_report(y_test_lin, y_pred_sm))

print("Matrice de confusion:")
conf_matrix_sm = confusion_matrix(y_test_lin, y_pred_sm)
print(conf_matrix_sm)

# Métriques binaires
if num_classes_sm == 2:
    TN, FP, FN, TP = conf_matrix_sm.ravel()
    print(f"\nTPR (Sensibilité): {TP/(TP+FN):.4f}")
    print(f"TNR (Spécificité): {TN/(TN+FP):.4f}")
    print(f"FPR: {FP/(FP+TN):.4f}")
    print(f"FNR: {FN/(FN+TP):.4f}")

# ---------------------------------------
# 3. RECHERCHE SYSTÉMATIQUE DES MEILLEURS HYPERPARAMÈTRES
# ---------------------------------------

print("\n" + "="*50)
print("=== RECHERCHE DES MEILLEURS HYPERPARAMÈTRES ===")
print("="*50)

# Grid Search pour Softmax
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['lbfgs', 'newton-cg'],
    'class_weight': [None, 'balanced']
}

grid_search = GridSearchCV(
    LogisticRegression(multi_class='multinomial', max_iter=2000),
    param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

grid_search.fit(X_train_lin, y_train_lin)

print(f"Meilleurs paramètres: {grid_search.best_params_}")
print(f"Meilleur score validation: {grid_search.best_score_:.4f}")

# Modèle avec les meilleurs paramètres
best_softmax = grid_search.best_estimator_
y_pred_best_sm = best_softmax.predict(X_test_lin)
accuracy_best_sm = accuracy_score(y_test_lin, y_pred_best_sm)

print(f"Accuracy avec meilleurs paramètres: {accuracy_best_sm:.4f}")

# ---------------------------------------
# 4. VISUALISATION AMÉLIORÉE
# ---------------------------------------

plt.figure(figsize=(18, 5))

# Confusion Matrix Linear Regression
plt.subplot(1, 4, 1)
plt.imshow(conf_matrix_lin, cmap='Blues')
plt.title(f"Régression Linéaire\nAccuracy: {accuracy_lin:.3f}")
plt.colorbar()
for i in range(conf_matrix_lin.shape[0]):
    for j in range(conf_matrix_lin.shape[1]):
        plt.text(j, i, conf_matrix_lin[i, j], ha='center', va='center', color='red')

# Confusion Matrix Softmax
plt.subplot(1, 4, 2)
plt.imshow(conf_matrix_sm, cmap='Oranges')
plt.title(f"Softmax Regression\nAccuracy: {accuracy_sm:.3f}")
plt.colorbar()
for i in range(conf_matrix_sm.shape[0]):
    for j in range(conf_matrix_sm.shape[1]):
        plt.text(j, i, conf_matrix_sm[i, j], ha='center', va='center', color='red')

# Confusion Matrix Softmax Optimisé
conf_matrix_best_sm = confusion_matrix(y_test_lin, y_pred_best_sm)
plt.subplot(1, 4, 3)
plt.imshow(conf_matrix_best_sm, cmap='Greens')
plt.title(f"Softmax Optimisé\nAccuracy: {accuracy_best_sm:.3f}")
plt.colorbar()
for i in range(conf_matrix_best_sm.shape[0]):
    for j in range(conf_matrix_best_sm.shape[1]):
        plt.text(j, i, conf_matrix_best_sm[i, j], ha='center', va='center', color='red')

# Accuracy comparison
plt.subplot(1, 4, 4)
models = ["Linear", "Softmax", "Softmax Opt"]
accuracies = [accuracy_lin, accuracy_sm, accuracy_best_sm]
colors = ['blue', 'orange', 'green']

bars = plt.bar(models, accuracies, alpha=0.7, color=colors)
plt.ylim(0, 1)
plt.title("Comparaison Accuracy")
for bar, acc in zip(bars, accuracies):
    plt.text(bar.get_x() + bar.get_width()/2, acc + 0.01, f"{acc:.3f}", 
             ha="center", va="bottom", fontweight='bold')

plt.tight_layout()
plt.show()

# ---------------------------------------
# 5. RÉSUMÉ COMPLET
# ---------------------------------------

print("\n" + "="*60)
print("RÉSUMÉ COMPARATIF DES MODÈLES AMÉLIORÉS")
print("="*60)

print(f"\nRégression Linéaire: Accuracy = {accuracy_lin:.4f}")
print(f"Softmax Regression : Accuracy = {accuracy_sm:.4f}")
print(f"Softmax Optimisé   : Accuracy = {accuracy_best_sm:.4f}")
print(f"AUC-ROC Softmax    : {auc_roc:.4f}")

print(f"\nAmélioration Softmax vs Linéaire: {(accuracy_sm - accuracy_lin):.4f}")
print(f"Amélioration Optimisé vs Base: {(accuracy_best_sm - accuracy_sm):.4f}")

# Analyse des erreurs
print(f"\n=== ANALYSE DES ERREURS ===")
print(f"Faux négatifs (danger médical): {conf_matrix_sm[1,0]}")
print(f"Faux positifs: {conf_matrix_sm[0,1]}")
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, LogisticRegression, Ridge, Lasso, RidgeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier, StackingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, precision_recall_curve, roc_curve
from sklearn.feature_selection import SelectKBest, f_classif, RFE
from sklearn.model_selection import cross_val_score, GridSearchCV, RandomizedSearchCV, train_test_split
from sklearn.preprocessing import StandardScaler, PowerTransformer, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import make_scorer
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# --------------------------------------------------
# 1. PRÉPARATION DES DONNÉES AVANCÉE
# --------------------------------------------------

print("=== PRÉPARATION AVANCÉE DES DONNÉES ===")

# Supposons que df et target_column sont définis
# Pour cet exemple, nous allons utiliser le dataset breast cancer
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
target_column = 'target'

# Inverser les labels pour que 0 = bénin, 1 = malin (comme dans votre contexte)
df['target'] = 1 - df['target']

# Vérifier et encoder les labels si nécessaire
if df[target_column].dtype == 'object':
    le = LabelEncoder()
    labels = le.fit_transform(df[target_column])
    print(f"Labels encodés: {le.classes_} -> {le.transform(le.classes_)}")
else:
    labels = df[target_column].values

features = df.drop(columns=[target_column]).values

print(f"Shape des features: {features.shape}")
print(f"Distribution des labels: {np.unique(labels, return_counts=True)}")

# Transformation avancée pour normalité
power_transformer = PowerTransformer(method='yeo-johnson')
features_transformed = power_transformer.fit_transform(features)

# Standardisation
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_transformed)

# Réduction de dimension avec PCA
pca = PCA(n_components=0.95)  # Garder 95% de la variance
features_pca = pca.fit_transform(features_scaled)

print(f"Dimensions originales: {features.shape[1]}")
print(f"Dimensions après PCA: {features_pca.shape[1]}")
print(f"Variance expliquée: {np.sum(pca.explained_variance_ratio_):.4f}")

# Split des données
X_train, X_test, y_train, y_test = train_test_split(
    features_pca, labels, test_size=0.25, stratify=labels, random_state=42
)

print(f"Train set: {X_train.shape[0]} échantillons")
print(f"Test set: {X_test.shape[0]} échantillons")

# --------------------------------------------------
# 2. MODÈLES AVANCÉS AVEC OPTIMISATION
# --------------------------------------------------

models = {}
predictions = {}
probabilities = {}

# 2.1 Ridge Classifier (CORRIGÉ)
print("\n" + "="*50)
print("RIDGE CLASSIFIER")
print("="*50)

ridge_params = {
    'alpha': [0.001, 0.01, 0.1, 1, 10, 100]
}

ridge_grid = GridSearchCV(RidgeClassifier(random_state=42), ridge_params, cv=5, scoring='accuracy')
ridge_grid.fit(X_train, y_train)
best_ridge = ridge_grid.best_estimator_
y_pred_ridge = best_ridge.predict(X_test)
accuracy_ridge = accuracy_score(y_test, y_pred_ridge)

print(f"Meilleurs paramètres Ridge: {ridge_grid.best_params_}")
print(f"Accuracy Ridge: {accuracy_ridge:.4f}")

# Calibration pour obtenir les probabilités
calibrated_ridge = CalibratedClassifierCV(best_ridge, method='sigmoid', cv=5)
calibrated_ridge.fit(X_train, y_train)
y_proba_ridge = calibrated_ridge.predict_proba(X_test)[:, 1]

models['Ridge Classifier'] = calibrated_ridge
predictions['Ridge Classifier'] = y_pred_ridge
probabilities['Ridge Classifier'] = y_proba_ridge

# 2.2 Softmax avec Optimisation Avancée
print("\n" + "="*50)
print("SOFTMAX AVEC OPTIMISATION COMPLÈTE")
print("="*50)

softmax_params = {
    'C': np.logspace(-3, 3, 10),
    'solver': ['lbfgs', 'newton-cg'],
    'class_weight': [None, 'balanced'],
    'max_iter': [2000]
}

softmax_grid = RandomizedSearchCV(
    LogisticRegression(multi_class='multinomial', random_state=42),
    softmax_params,
    n_iter=20,
    cv=5,
    scoring='accuracy',
    random_state=42,
    n_jobs=-1
)

softmax_grid.fit(X_train, y_train)
best_softmax = softmax_grid.best_estimator_
y_pred_softmax = best_softmax.predict(X_test)
y_proba_softmax = best_softmax.predict_proba(X_test)[:, 1]
accuracy_softmax = accuracy_score(y_test, y_pred_softmax)

print(f"Meilleurs paramètres Softmax: {softmax_grid.best_params_}")
print(f"Accuracy Softmax: {accuracy_softmax:.4f}")

models['Softmax'] = best_softmax
predictions['Softmax'] = y_pred_softmax
probabilities['Softmax'] = y_proba_softmax

# 2.3 Random Forest (CORRIGÉ)
print("\n" + "="*50)
print("RANDOM FOREST")
print("="*50)

rf_params = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'class_weight': ['balanced', None]
}

rf_grid = RandomizedSearchCV(
    RandomForestClassifier(random_state=42),
    rf_params,
    n_iter=10,
    cv=5,
    scoring='accuracy',
    random_state=42,
    n_jobs=-1
)

rf_grid.fit(X_train, y_train)
best_rf = rf_grid.best_estimator_
y_pred_rf = best_rf.predict(X_test)
y_proba_rf = best_rf.predict_proba(X_test)[:, 1]
accuracy_rf = accuracy_score(y_test, y_pred_rf)

print(f"Meilleurs paramètres Random Forest: {rf_grid.best_params_}")
print(f"Accuracy Random Forest: {accuracy_rf:.4f}")

models['Random Forest'] = best_rf
predictions['Random Forest'] = y_pred_rf
probabilities['Random Forest'] = y_proba_rf

# 2.4 Gradient Boosting
print("\n" + "="*50)
print("GRADIENT BOOSTING")
print("="*50)

gb_params = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 4],
    'subsample': [0.8, 0.9]
}

gb_grid = RandomizedSearchCV(
    GradientBoostingClassifier(random_state=42),
    gb_params,
    n_iter=8,
    cv=5,
    scoring='accuracy',
    random_state=42
)

gb_grid.fit(X_train, y_train)
best_gb = gb_grid.best_estimator_
y_pred_gb = best_gb.predict(X_test)
y_proba_gb = best_gb.predict_proba(X_test)[:, 1]
accuracy_gb = accuracy_score(y_test, y_pred_gb)

print(f"Meilleurs paramètres Gradient Boosting: {gb_grid.best_params_}")
print(f"Accuracy Gradient Boosting: {accuracy_gb:.4f}")

models['Gradient Boosting'] = best_gb
predictions['Gradient Boosting'] = y_pred_gb
probabilities['Gradient Boosting'] = y_proba_gb

# 2.5 SVM
print("\n" + "="*50)
print("SVM")
print("="*50)

svm_params = {
    'C': [0.1, 1, 10],
    'gamma': ['scale', 'auto'],
    'kernel': ['rbf', 'linear']
}

svm_grid = RandomizedSearchCV(
    SVC(probability=True, random_state=42),
    svm_params,
    n_iter=6,
    cv=5,
    scoring='accuracy',
    random_state=42
)

svm_grid.fit(X_train, y_train)
best_svm = svm_grid.best_estimator_
y_pred_svm = best_svm.predict(X_test)
y_proba_svm = best_svm.predict_proba(X_test)[:, 1]
accuracy_svm = accuracy_score(y_test, y_pred_svm)

print(f"Meilleurs paramètres SVM: {svm_grid.best_params_}")
print(f"Accuracy SVM: {accuracy_svm:.4f}")

models['SVM'] = best_svm
predictions['SVM'] = y_pred_svm
probabilities['SVM'] = y_proba_svm

# 2.6 XGBoost pour améliorer la sensibilité
print("\n" + "="*50)
print("XGBOOST (POUR AMÉLIORER LA SENSIBILITÉ)")
print("="*50)

try:
    from xgboost import XGBClassifier
    
    # Calcul du poids pour gérer le déséquilibre des classes
    scale_pos_weight = len(y_train[y_train==0]) / len(y_train[y_train==1])
    
    xgb_params = {
        'n_estimators': [100, 200],
        'max_depth': [3, 4, 5],
        'learning_rate': [0.01, 0.1],
        'scale_pos_weight': [scale_pos_weight, 1]
    }
    
    xgb_grid = RandomizedSearchCV(
        XGBClassifier(random_state=42, eval_metric='logloss'),
        xgb_params,
        n_iter=8,
        cv=5,
        scoring='accuracy',
        random_state=42
    )
    
    xgb_grid.fit(X_train, y_train)
    best_xgb = xgb_grid.best_estimator_
    y_pred_xgb = best_xgb.predict(X_test)
    y_proba_xgb = best_xgb.predict_proba(X_test)[:, 1]
    accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
    
    print(f"Meilleurs paramètres XGBoost: {xgb_grid.best_params_}")
    print(f"Accuracy XGBoost: {accuracy_xgb:.4f}")
    
    models['XGBoost'] = best_xgb
    predictions['XGBoost'] = y_pred_xgb
    probabilities['XGBoost'] = y_proba_xgb
    
except ImportError:
    print("XGBoost non installé, poursuite sans XGBoost")

# --------------------------------------------------
# 3. ENSEMBLE LEARNING
# --------------------------------------------------

print("\n" + "="*50)
print("MODÈLES ENSEMBLE")
print("="*50)

# Préparer les estimateurs pour l'ensemble
estimators_for_ensemble = [
    ('softmax', best_softmax),
    ('rf', best_rf),
    ('gb', best_gb),
    ('svm', best_svm)
]

# Voting Classifier
voting_clf = VotingClassifier(
    estimators=estimators_for_ensemble,
    voting='soft'
)

voting_clf.fit(X_train, y_train)
y_pred_voting = voting_clf.predict(X_test)
y_proba_voting = voting_clf.predict_proba(X_test)[:, 1]
accuracy_voting = accuracy_score(y_test, y_pred_voting)

print(f"Accuracy Voting Ensemble: {accuracy_voting:.4f}")

models['Voting Ensemble'] = voting_clf
predictions['Voting Ensemble'] = y_pred_voting
probabilities['Voting Ensemble'] = y_proba_voting

# Stacking Classifier
stacking_clf = StackingClassifier(
    estimators=estimators_for_ensemble,
    final_estimator=LogisticRegression(),
    cv=5
)

stacking_clf.fit(X_train, y_train)
y_pred_stacking = stacking_clf.predict(X_test)
y_proba_stacking = stacking_clf.predict_proba(X_test)[:, 1]
accuracy_stacking = accuracy_score(y_test, y_pred_stacking)

print(f"Accuracy Stacking: {accuracy_stacking:.4f}")

models['Stacking'] = stacking_clf
predictions['Stacking'] = y_pred_stacking
probabilities['Stacking'] = y_proba_stacking

# --------------------------------------------------
# 4. ANALYSE DES PERFORMANCES
# --------------------------------------------------

print("\n" + "="*60)
print("COMPARAISON DES PERFORMANCES")
print("="*60)

# Calcul des accuracies
results = {}
for model_name in models:
    if model_name in predictions:
        results[model_name] = accuracy_score(y_test, predictions[model_name])

print("\nClassement des modèles par performance:")
print("-" * 50)
for model_name, accuracy in sorted(results.items(), key=lambda x: x[1], reverse=True):
    print(f"{model_name:20}: {accuracy:.4f} ({accuracy*100:.2f}%)")

# Meilleur modèle
best_model_name = max(results, key=results.get)
best_model = models[best_model_name]
best_pred = predictions[best_model_name]
best_proba = probabilities[best_model_name] if best_model_name in probabilities else None

print(f"\n🌟 MEILLEUR MODÈLE: {best_model_name}")
print(f"📊 Accuracy: {results[best_model_name]:.4f}")

# Validation croisée du meilleur modèle (CORRIGÉE)
try:
    if hasattr(best_model, 'predict_proba') or hasattr(best_model, 'predict'):
        cv_scores = cross_val_score(best_model, X_train, y_train, cv=5, scoring='accuracy')
        print(f"🔍 Validation croisée (5-fold): {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")
    else:
        print("🔍 Validation croisée: Modèle non compatible avec cross_val_score")
except Exception as e:
    print(f"🔍 Validation croisée échouée: {e}")

# --------------------------------------------------
# 5. VISUALISATIONS AVANCÉES
# --------------------------------------------------

plt.figure(figsize=(20, 6))

# Graphique 1: Comparaison des accuracy
plt.subplot(1, 3, 1)
models_names = list(results.keys())
accuracies = list(results.values())
colors = ['red' if 'Ridge' in name else 'blue' if 'Softmax' in name else 'green' for name in models_names]

bars = plt.bar(models_names, accuracies, color=colors, alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.ylim(0.9, 1.0)
plt.title('Comparaison des Accuracies des Modèles', fontsize=14, fontweight='bold')
plt.ylabel('Accuracy')
plt.grid(True, alpha=0.3)

for bar, acc in zip(bars, accuracies):
    plt.text(bar.get_x() + bar.get_width()/2, acc + 0.005, f'{acc:.3f}', 
             ha='center', va='bottom', fontweight='bold')

# Graphique 2: Matrice de confusion du meilleur modèle
plt.subplot(1, 3, 2)
conf_matrix = confusion_matrix(y_test, best_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Bénin', 'Malin'], yticklabels=['Bénin', 'Malin'])
plt.title(f'Matrice de Confusion - {best_model_name}\nAccuracy: {results[best_model_name]:.4f}', 
          fontsize=12, fontweight='bold')
plt.xlabel('Prédit')
plt.ylabel('Réel')

# Graphique 3: Courbe ROC pour tous les modèles
plt.subplot(1, 3, 3)
for model_name in probabilities:
    if model_name in probabilities and probabilities[model_name] is not None:
        try:
            fpr, tpr, _ = roc_curve(y_test, probabilities[model_name])
            auc_score = roc_auc_score(y_test, probabilities[model_name])
            plt.plot(fpr, tpr, label=f'{model_name} (AUC = {auc_score:.3f})', linewidth=2)
        except Exception as e:
            print(f"Erreur ROC pour {model_name}: {e}")

plt.plot([0, 1], [0, 1], 'k--', alpha=0.5, label='Aléatoire')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Courbes ROC - Comparaison des Modèles', fontsize=12, fontweight='bold')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# --------------------------------------------------
# 6. ANALYSE DÉTAILLÉE DU MEILLEUR MODÈLE
# --------------------------------------------------

print("\n" + "="*60)
print("ANALYSE DÉTAILLÉE DU MEILLEUR MODÈLE")
print("="*60)

print(f"\n📋 Modèle: {best_model_name}")
print(f"🎯 Accuracy finale: {results[best_model_name]:.4f}")

# Rapport de classification détaillé
print("\n📊 Rapport de classification:")
print(classification_report(y_test, best_pred, target_names=['Bénin', 'Malin']))

# Métriques importantes pour le cancer
TN, FP, FN, TP = conf_matrix.ravel()
print(f"\n⚕️  MÉTRIQUES MÉDICALES IMPORTANTES:")
print(f"   Sensibilité (Recall Malin): {TP/(TP+FN):.4f}  ← CRITIQUE")
print(f"   Spécificité (Recall Bénin): {TN/(TN+FP):.4f}")
print(f"   Précision Malin: {TP/(TP+FP):.4f}" if (TP+FP) > 0 else "   Précision Malin: Indéfinie")
print(f"   Faux Négatifs (cas dangereux): {FN}  ← À MINIMISER")
print(f"   Faux Positifs: {FP}")
print(f"   Score F1: {2*TP/(2*TP + FP + FN):.4f}" if (2*TP + FP + FN) > 0 else "   Score F1: Indéfini")

# Calcul AUC-ROC si disponible
if best_proba is not None:
    try:
        auc_roc = roc_auc_score(y_test, best_proba)
        print(f"   AUC-ROC: {auc_roc:.4f}")
    except:
        print("   AUC-ROC: Non calculable")

# --------------------------------------------------
# 7. OPTIMISATION DU SEUIL DE DÉCISION POUR RÉDUIRE LES FAUX NÉGATIFS
# --------------------------------------------------

print("\n" + "="*60)
print("OPTIMISATION DU SEUIL DE DÉCISION POUR RÉDUIRE LES FAUX NÉGATIFS")
print("="*60)

if best_proba is not None:
    # Recherche du seuil optimal avec focus sur la réduction des faux négatifs
    precisions, recalls, thresholds = precision_recall_curve(y_test, best_proba)
    
    # Calcul des F1-scores
    f1_scores = 2 * (precisions * recalls) / (precisions + recalls)
    f1_scores = np.nan_to_num(f1_scores)
    
    # Seuil pour maximiser F1
    if len(f1_scores) > 0:
        best_threshold_f1_idx = np.argmax(f1_scores[:-1])
        best_threshold_f1 = thresholds[best_threshold_f1_idx]
    
    # Seuil pour cibler une sensibilité élevée (≥98%)
    target_sensitivity = 0.98
    sensitivity_threshold_idx = np.where(recalls[:-1] >= target_sensitivity)[0]
    
    if len(sensitivity_threshold_idx) > 0:
        best_threshold_sensitivity = thresholds[sensitivity_threshold_idx[0]]
    else:
        best_threshold_sensitivity = 0.5
    
    print(f"🔧 Seuil par défaut: 0.5")
    print(f"🎯 Meilleur seuil pour F1-score: {best_threshold_f1:.4f}")
    print(f"🎯 Seuil pour sensibilité ≥98%: {best_threshold_sensitivity:.4f}")
    
    # Application du seuil optimisé pour sensibilité
    y_pred_optimized = (best_proba > best_threshold_sensitivity).astype(int)
    accuracy_optimized = accuracy_score(y_test, y_pred_optimized)
    
    # Métriques avec seuil optimisé
    TN_opt, FP_opt, FN_opt, TP_opt = confusion_matrix(y_test, y_pred_optimized).ravel()
    sensitivity_opt = TP_opt / (TP_opt + FN_opt)
    specificity_opt = TN_opt / (TN_opt + FP_opt)
    
    print(f"\n📈 RÉSULTATS AVEC SEUIL OPTIMISÉ (Sensibilité ≥98%):")
    print(f"   Accuracy: {accuracy_optimized:.4f}")
    print(f"   Sensibilité: {sensitivity_opt:.4f}")
    print(f"   Spécificité: {specificity_opt:.4f}")
    print(f"   Faux Négatifs: {FN_opt} (amélioration: {FN - FN_opt})")
    print(f"   Faux Positifs: {FP_opt}")

# --------------------------------------------------
# 8. ANALYSE DES ERREURS
# --------------------------------------------------

print("\n" + "="*60)
print("ANALYSE DES ÉCHANTILLONS MAL CLASSÉS")
print("="*60)

# Identifier les échantillons mal classés
misclassified_idx = np.where(y_test != best_pred)[0]

if len(misclassified_idx) > 0:
    print(f"Nombre d'échantillons mal classés: {len(misclassified_idx)}")
    
    # Analyser les caractéristiques des erreurs
    error_analysis = []
    for idx in misclassified_idx:
        true_label = y_test[idx]
        pred_label = best_pred[idx]
        proba = best_proba[idx] if best_proba is not None else None
        
        error_type = "Faux Négatif" if true_label == 1 and pred_label == 0 else "Faux Positif"
        error_analysis.append({
            'index': idx,
            'true_label': true_label,
            'predicted_label': pred_label,
            'error_type': error_type,
            'probability': proba
        })
    
    # Afficher l'analyse des erreurs
    error_df = pd.DataFrame(error_analysis)
    print("\nDétail des erreurs:")
    print(error_df[['error_type', 'true_label', 'predicted_label', 'probability']])
    
    # Statistiques des erreurs
    fn_count = len([e for e in error_analysis if e['error_type'] == 'Faux Négatif'])
    fp_count = len([e for e in error_analysis if e['error_type'] == 'Faux Positif'])
    
    print(f"\n📊 RÉCAPITULATIF DES ERREURS:")
    print(f"   Faux Négatifs (Cancer non détecté): {fn_count}")
    print(f"   Faux Positifs: {fp_count}")
    
else:
    print("🎉 Aucun échantillon mal classé!")

# --------------------------------------------------
# 9. RÉSUMÉ FINAL ET RECOMMANDATIONS
# --------------------------------------------------

print("\n" + "="*60)
print("RÉSUMÉ FINAL ET RECOMMANDATIONS")
print("="*60)

print(f"\n✅ MEILLEUR MODÈLE: {best_model_name}")
print(f"✅ ACCURACY: {results[best_model_name]:.4f} ({results[best_model_name]*100:.2f}%)")

if best_proba is not None:
    try:
        auc_roc = roc_auc_score(y_test, best_proba)
        print(f"✅ AUC-ROC: {auc_roc:.4f}")
    except:
        pass

print(f"✅ SENSIBILITÉ: {TP/(TP+FN):.4f}")
print(f"✅ SPÉCIFICITÉ: {TN/(TN+FP):.4f}")
print(f"✅ FAUX NÉGATIFS: {FN}")

# Recommandations basées sur les résultats
if FN == 0:
    print("🎉 EXCELLENT! Aucun faux négatif - tous les cas de cancer détectés!")
elif FN <= 1:
    print("👍 TRÈS BIEN! Très peu de faux négatifs")
else:
    print("⚠️  Attention: plusieurs faux négatifs détectés - optimisation recommandée")

print(f"\n📈 PERFORMANCE GLOBALE:")
print(f"   Le modèle {best_model_name} montre une performance exceptionnelle")
print(f"   avec une précision de {results[best_model_name]*100:.2f}%.")

print(f"\n🔧 RECOMMANDATIONS:")
if FN > 0:
    print("   1. Utiliser le seuil optimisé pour réduire les faux négatifs")
    print("   2. Explorer XGBoost avec focal loss pour améliorer la sensibilité")
    print("   3. Considérer l'augmentation des données pour les cas difficiles")
else:
    print("   1. Le modèle est optimal pour la détection du cancer")
    print("   2. Maintenir le monitoring des performances sur de nouvelles données")
    print("   3. Considérer le déploiement en environnement de production")

print(f"\n🎯 POUR LA SÉCURITÉ DES PATIENTS:")
print("   Priorité absolue: Minimiser les faux négatifs (cancers non détectés)")
print("   Les faux positifs sont moins critiques mais à surveiller")

print(f"\n💡 PROCHAINES ÉTAPES:")
print("   - Validation sur un dataset externe")
print("   - Tests d'robustesse avec différentes distributions")
print("   - Déploiement progressif avec monitoring continu")
```
